```python

# Decorators: enhance the base function
import time

def timer(base_fn):
  def enhanced_fn():
    start_time = time.time()
    base_fn()
    stop_time = time.time()
    print(f"Time taken: {stop_time - start_time} seconds")
  return enhanced_fn

@timer
def brewtea():
  print("Making Tea")
  time.sleep(1)
  print("Tea is ready")

brewtea()
```
